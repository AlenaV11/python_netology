```python
# Задание 3
n = int(input('Input '))

def fibonacci():
    a, b = 1, 1
    for i in range(n):
        yield a
        a, b = b, a + b

data = list(fibonacci())
print(data)

summ = sum(data)
print('Сумма первых', n, 'чисел Фибона́ччи — ', summ)  
```
